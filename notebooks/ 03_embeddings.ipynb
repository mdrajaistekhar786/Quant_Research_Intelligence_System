```python
import json
import torch

print("CUDA Available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
#Loading Chunks
INPUT_PATH = "../data/processed_papers/chunked_papers.json"

with open(
    INPUT_PATH,
    "r",
    encoding="utf-8"
) as f:
    chunks = json.load(f)

print("Total Chunks:", len(chunks))
```

```python
chunks[0]
```

```python
import sys
print(sys.executable)
```

```python
from sentence_transformers import SentenceTransformer

print("Installed Successfully")
```

```python
texts = [chunk["text"] for chunk in chunks]

print("Total texts:", len(texts))
print(texts[0][:300])
```

```python
from sentence_transformers import SentenceTransformer

MODEL_NAME = "BAAI/bge-small-en-v1.5"

model = SentenceTransformer(
    MODEL_NAME,
    device="cuda"
)

print("Model Loaded Successfully")
```

```python
embeddings = model.encode(
    texts,
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True
)

print(embeddings.shape)
```

```python
print(embeddings[0][:10])
print(len(embeddings[0]))
```

```python
for i, chunk in enumerate(chunks):
    chunk["embedding"] = embeddings[i].tolist()

print("Embeddings attached successfully")
```

```python
chunks[0].keys()
```

```python
import json

OUTPUT_PATH = "../data/processed_papers/embedded_chunks.json"

with open(
    OUTPUT_PATH,
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        chunks,
        f,
        ensure_ascii=False
    )

print("Embedded chunks saved successfully")
```

```python
with open(
    OUTPUT_PATH,
    "r",
    encoding="utf-8"
) as f:

    embedded_chunks = json.load(f)

print(len(embedded_chunks))
print(len(embedded_chunks[0]["embedding"]))
```
