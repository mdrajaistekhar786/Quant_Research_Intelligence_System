```python
import os
import re
import json
import pandas as pd

from pypdf import PdfReader
from tqdm import tqdm
```

```python
PDF_FOLDER = "../data/raw_papers"

pdf_files = [
    f for f in os.listdir(PDF_FOLDER)
    if f.endswith(".pdf")
]

print(f"Found {len(pdf_files)} PDFs")
print(pdf_files)
```

```python
def extract_pdf_text(pdf_path):

    text = ""

    try:
        reader = PdfReader(pdf_path)

        for page in reader.pages:

            page_text = page.extract_text()

            if page_text:
                text += page_text + "\n"

    except Exception as e:
        print(f"Error: {pdf_path}")
        print(e)

    return text
```

```python
sample_pdf = os.path.join(
    PDF_FOLDER,
    pdf_files[0]
)

text = extract_pdf_text(sample_pdf)

print(text[:3000])
```

```python
def extract_metadata(text, filename):

    lines = text.split("\n")

    title = ""

    for line in lines[:15]:

        line = line.strip()

        if len(line) > 20:
            title = line
            break

    return {
        "filename": filename,
        "title": title,
        "text": text
    }
```

```python
all_papers = []

for pdf_file in tqdm(pdf_files):

    pdf_path = os.path.join(
        PDF_FOLDER,
        pdf_file
    )

    text = extract_pdf_text(pdf_path)

    paper = extract_metadata(
        text,
        pdf_file
    )

    all_papers.append(paper)
```

```python
print(f"Total papers: {len(all_papers)}")

print("\n")

print(all_papers[0]["title"])

print("\n")

print(all_papers[0]["filename"])
```

```python
OUTPUT_PATH = "../data/processed_papers/papers_metadata.json"

with open(
    OUTPUT_PATH,
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        all_papers,
        f,
        indent=4,
        ensure_ascii=False
    )

print("Saved successfully")
```
