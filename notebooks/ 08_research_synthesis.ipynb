```python
import json

with open(
    "../data/processed_papers/chunked_papers.json",
    "r",
    encoding="utf-8"
) as f:
    chunks = json.load(f)

print(len(chunks))
```

```python
papers = {}

for chunk in chunks:

    title = chunk["paper_title"]

    if title not in papers:
        papers[title] = []

    papers[title].append(
        chunk["text"]
    )

print(len(papers))
```

```python
paper_summaries = {}

for title, paper_chunks in papers.items():

    combined_text = "\n".join(
        paper_chunks[:5]
    )

    paper_summaries[title] = combined_text
```

```python
for title in list(paper_summaries.keys())[:3]:

    print("\n")
    print(title)
    print("-" * 80)

    print(
        paper_summaries[title][:500]
    )
```

```python
def find_relevant_papers(keyword):

    relevant = {}

    for title, summary in paper_summaries.items():

        if keyword.lower() in summary.lower():

            relevant[title] = summary

    return relevant
```

```python
relevant_papers = find_relevant_papers(
    "stochastic volatility"
)

print(
    f"Found {len(relevant_papers)} papers"
)

for title in relevant_papers.keys():

    print("\n", title)
```

```python
def compare_papers(keyword):

    papers = find_relevant_papers(keyword)

    print("\nTOPIC:")
    print(keyword)

    print("\nRELEVANT PAPERS:")
    print("=" * 80)

    for title in papers:

        print("\n•", title)

    return papers
```

```python
compare_papers(
    "stochastic volatility"
)
```

```python
research_corpus = compare_papers(
    "stochastic volatility"
)

print(
    "\nNumber of relevant papers:",
    len(research_corpus)
)
```

```python
def generate_research_summary(keyword):

    papers = find_relevant_papers(keyword)

    print("\nRESEARCH SYNTHESIS")
    print("=" * 80)

    print("\nTOPIC:")
    print(keyword)

    print("\nNUMBER OF PAPERS:")
    print(len(papers))

    print("\nPAPERS:")

    for i, title in enumerate(papers.keys(), start=1):

        print(f"\n{i}. {title}")

        summary = papers[title][:500]

        print("\nKey Content:")

        print(summary)
```

```python
generate_research_summary(
    "stochastic volatility"
)
```

```python
def identify_research_gap(keyword):

    papers = find_relevant_papers(keyword)

    print("\nPOTENTIAL RESEARCH GAPS")
    print("=" * 80)

    print(
        "Manual review required:"
    )

    print(
        "- Compare methodologies"
    )

    print(
        "- Compare assumptions"
    )

    print(
        "- Compare datasets"
    )

    print(
        "- Compare performance metrics"
    )

    print(
        "- Look for unexplored combinations"
    )
```

```python
identify_research_gap(
    "stochastic volatility"
)
```

```python
def compare_papers(keyword):

    papers = find_relevant_papers(keyword)

    if len(papers) < 2:
        print("Need at least 2 papers")
        return

    paper_list = list(papers.items())

    print("\nPAPER 1")
    print("=" * 80)
    print(paper_list[0][0])
    print("\n")
    print(paper_list[0][1][:1000])

    print("\n\nPAPER 2")
    print("=" * 80)
    print(paper_list[1][0])
    print("\n")
    print(paper_list[1][1][:1000])
```

```python
compare_papers(
    "stochastic volatility"
)
```

```python
def create_research_report(keyword):

    papers = find_relevant_papers(keyword)

    report = {
        "topic": keyword,
        "number_of_papers": len(papers),
        "paper_titles": list(papers.keys())
    }

    return report
```

```python
report = create_research_report(
    "stochastic volatility"
)

print(report)
```
