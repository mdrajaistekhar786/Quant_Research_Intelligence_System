```python
def split_into_sections(text):

    pattern = r'\n\d+\s+[A-Z][A-Za-z\s]+\n'

    sections = re.split(pattern, text)

    headers = re.findall(pattern, text)

    return headers, sections
```

```python
headers, sections = split_into_sections(
    papers[0]["text"]
)

print(headers[:10])

print("\n")

print(f"Number of sections found: {len(sections)}")
```

```python
def clean_text(text):

    text = re.sub(r'\s+', ' ', text)

    text = text.replace('\x00', ' ')

    text = text.strip()

    return text
```

```python
cleaned_text = clean_text(
    papers[0]["text"]
)

print(cleaned_text[:1000])
```

```python
def create_chunks(text,
                  chunk_size=1000,
                  overlap=200):

    chunks = []

    start = 0

    while start < len(text):

        end = start + chunk_size

        chunk = text[start:end]

        chunks.append(chunk)

        start += chunk_size - overlap

    return chunks
```

```python
chunks = create_chunks(cleaned_text)

print(f"Total chunks: {len(chunks)}")
```

```python
all_chunks = []

global_chunk_id = 0

for paper in papers:

    cleaned_text = clean_text(
        paper["text"]
    )

    chunks = create_chunks(
        cleaned_text
    )

    for local_idx, chunk in enumerate(chunks):

        chunk_data = {
            "global_chunk_id": global_chunk_id,
            "local_chunk_id": local_idx,
            "paper_title": paper["title"],
            "filename": paper["filename"],
            "text": chunk
        }

        all_chunks.append(chunk_data)

        global_chunk_id += 1
```

```python
print(f"Total chunks created: {len(all_chunks)}")
```

```python
OUTPUT_PATH = "../data/processed_papers/chunked_papers.json"
```

```python
with open(
    OUTPUT_PATH,
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        all_chunks,
        f,
        indent=4,
        ensure_ascii=False
    )

print("Chunks saved successfully")
```

```python
with open(
    OUTPUT_PATH,
    "r",
    encoding="utf-8"
) as f:

    loaded_chunks = json.load(f)

print(len(loaded_chunks))
```
