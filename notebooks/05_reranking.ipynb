```python
from sentence_transformers import CrossEncoder
import numpy as np
```

```python
RERANK_MODEL = "cross-encoder/ms-marco-MiniLM-L-6-v2"

reranker = CrossEncoder(
    RERANK_MODEL,
    device="cuda"
)

print("Cross Encoder Loaded")
```

```python
query = "stochastic volatility"

passage = """
Stochastic volatility models assume that volatility
changes randomly over time and follows its own process.
"""

score = reranker.predict(
    [(query, passage)]
)

print(score)
```
